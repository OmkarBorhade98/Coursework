### Loading from Remote URL

```python
url = "https://github.com/crux82/squad-it/raw/master/"
data_files ={
    'train': url+"SQuAD_it-train.json.gz",
    'test': url+"SQuAD_it-test.json.gz"
}
remote_data = load_dataset('json', data_files= data_files, field = 'data')
remote_data
```
