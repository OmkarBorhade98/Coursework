```python
from transformers import pipeline
```

## sentiment-analysis
Predicts positive/negative sentiment of a sentence or list of sentences

```python
# input sentence
classifier = pipeline('sentiment-analysis')
classifier('Old MacDonald had a farm')
```

```python
# input list of sentences
classifier([
    'New year new me',
    'Ah shit, here we go again',
    'All you have to do was to follow a damn train CJ!',
    'Do or do not, there is no try',
    'No means No!'
])
```

## zero-shot-classification
Classifies input sequence amoung the input list of candidate labels

```python
classifier = pipeline('zero-shot-classification')
classifier('My code is better than yours',
           candidate_labels = ['web developer', 'software engineer',
                               'hacker', 'cricketer', 'politician', 'encrypter'])
```

## text-generation
Generates Text

```python
generator = pipeline("text-generation")
generator("When the snows fall and the white winds blow, the lone wolf dies,")
```

```python
# text generation using distilgpt2 model
generator = pipeline('text-generation', model = 'distilgpt2')
generator(
    'When the snows fall and the white winds blow, the lone wolf dies',
    num_return_sequences = 2,
    max_length = 50
)
```

## fill-mask
Predicts and fills mask

```python
unmasker = pipeline('fill-mask')
unmasker('Ah, shit! Here we <mask> again',top_k =5 )
```

## ner
Named Entity Recongnition

```python
nlp = pipeline("ner", aggregation_strategy="simple")

sequence = "Hugging Face Inc. is a company based in New York City. Its headquarters are in DUMBO, therefore very" \
           "close to the Manhattan Bridge which is visible from the window."

print(nlp(sequence))
```

## question-answering
Answers questions based on given context

```python
answer = pipeline('question-answering')
answer(question="What is my purpose?",
    context="My name is Omkar and I work as a Software Engineer and AI devolaper")
```
